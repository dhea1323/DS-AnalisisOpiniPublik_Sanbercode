```python
def multiple_tweet(username, n_count, header):
    hasil = api.user_timeline(id = username , count = n_count)
    list_hasil = []
    for tweet in hasil:
        tweets = tweet.text
        list_hasil.append(tweets)
    return pd.DataFrame(list_hasil, columns = [header])
```
