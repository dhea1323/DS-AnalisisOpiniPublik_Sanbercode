```python
import requests
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory
from bs4 import BeautifulSoup
from urllib.request import urlopen
import pandas as pd
import re
from nltk.corpus import stopwords
import nltk
import string
```

```python
url = "https://nasional.kompas.com/read/2020/12/17/09084331/alasan-jokowi-putuskan-vaksin-covid-19-digratiskan-untuk-masyarakat?page=all#page2"
html = urlopen(url)
data = BeautifulSoup(html, 'html.parser')
```

# NLP dan Sastrawi

```python
#halaman 1
h1 = data.findAll("div", {"data-io-article-url":"https://nasional.kompas.com/read/2020/12/17/09084331/alasan-jokowi-putuskan-vaksin-covid-19-digratiskan-untuk-masyarakat"})[0]
p = h1.find_all("p")
hasil = []
for i in range (1,len(p)):
    isi = p[i].get_text()
    hasil.append(isi)
df = (pd.Series(hasil)).copy()
df
```

```python
def case_folding_stopword(data):
    new_data = data.str.lower().str.replace(r"\d+", "").str.replace(r'[^\w\s]', '').str.strip()
    nltk.download('stopwords')
    indonesia_s = stopwords.words('indonesian')
    querywords = new_data.str.split()
    resultwords = [word for word in querywords if word not in indonesia_s]
    result = (pd.Series(resultwords)).str.join(' ')
    return result  
```

```python
new_df =(case_folding_stopword(df))
new_df
```

```python
factory = StemmerFactory()
stemmer = factory.create_stemmer()

hasil_stem = []
for j in range(len(new_df)):
    output   = stemmer.stem(new_df[j])
    hasil_stem.append(output)
```

```python
topik = 'Vaksin Covid19 Gratis'
link = 'https://nasional.kompas.com/read/2020/12/17/09084331/alasan-jokowi-putuskan-vaksin-covid-19-digratiskan-untuk-masyarakat?page=all#page2'
clean_df = pd.DataFrame(hasil_stem, columns = ["Clean Dataframe"])
raw_df = pd.DataFrame(hasil, columns = ["Raw Dataframe"])
df_save = pd.concat([raw_df,clean_df], axis = 1)
df_save.insert(0,'Topik',topik)
df_save.insert(0, 'Link', link)
df_save.drop(df_save.index[[6]])
```

```python
df_save.to_csv('dhea1323_tugas11.csv', index = False)
```
